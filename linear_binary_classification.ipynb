```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.linspace(-10, 10, 1000)
ax.plot(x, 1 / (1 + np.exp(-x)))
ax.set_xlabel('$a$')
ax.set_ylabel('$\sigma(a)$')
ax.grid()
plt.show()
```

```python
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
```

```python
sigmoid(0)
```

```python
sigmoid(3), sigmoid(-3)
```
